```
# Scrapping job content from the official page

from langchain_community.document_loaders.web_base import WebBaseLoader
loader = WebBaseLoader("https://jobs.nike.com/job/R-49841?from=job%20search%20funnel")
page_data = loader.load().pop().page_content
```

```
print(page_data)
```

```
# prompting the llama model to use that scrapped content and extract the use full things for me

from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template(
    """
    ##SCRAPED TEXT FROM WEBSITE:
    {page_data}
    ### INSTRUCTION:
    The Scraped text is from the career's page of a website.
    Your job is to extract the job postings and return them in json format
    containing following kets: 'role', 'experience', 'skills' and 'description'.
    Only return the valid JSON.
    ### VAILD JSON (NO PREAMBLE)

    """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input = {"page_data": page_data})
print(res.content)
print(type(res.content))
```

```
# The scrapped content was in str, so converting it into json type which will make it easier to transfer

from langchain_core.output_parsers import JsonOutputParser
jason_parser = JsonOutputParser()
json_res = jason_parser.parse(res.content)
print(json_res)
print(type(json_res))
```
